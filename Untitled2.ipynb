```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# 1. GENERATE ADVANCED E-COMMERCE DATA
np.random.seed(42)
months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]
data = {
    "Month": np.random.choice(months, 200),
    "Category": np.random.choice(
        ["Smartphones", "Laptops", "Audio Gear", "Wearables"], 200, p=[0.4, 0.3, 0.2, 0.1]
    ),
    "Sales_USD": np.random.randint(500, 5000, 200),
    "Profit_USD": np.random.randint(-200, 1500, 200),
    "Segment": np.random.choice(
        ["Consumer", "Corporate", "Home Office"], 200
    ),
}
df = pd.DataFrame(data)
df["Month"] = pd.Categorical(df["Month"], categories=months, ordered=True)

# 2. SET MODERN COLOR PALETTES & THEME
sns.set_theme(style="darkgrid")
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["figure.facecolor"] = "#1e1e24"
plt.rcParams["axes.facecolor"] = "#2d2d34"
plt.rcParams["text.color"] = "white"
plt.rcParams["axes.labelcolor"] = "white"
plt.rcParams["xtick.color"] = "white"
plt.rcParams["ytick.color"] = "white"

# Create Canvas Layout (2x2 Structure)
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle(
    "CODEALPHA TASK 3: GLOBAL SALES INFOGRAPHIC DASHBOARD",
    fontsize=22,
    fontweight="bold",
    color="#00adb5",
    y=0.98,
)

# CHART 1: Area Chart For Monthly Performance Trend
df_monthly = df.groupby("Month", observed=False)["Sales_USD"].sum().reset_index()
axes[0, 0].fill_between(
    df_monthly["Month"],
    df_monthly["Sales_USD"],
    color="#00adb5",
    alpha=0.4,
)
sns.lineplot(
    ax=axes[0, 0],
    data=df_monthly,
    x="Month",
    y="Sales_USD",
    color="#00adb5",
    linewidth=3,
    marker="o",
    markersize=8,
)
axes[0, 0].set_title(
    "📦 Monthly Gross Revenue Trajectory", fontsize=14, fontweight="bold", color="#f8f9fa"
)
axes[0, 0].set_ylabel("Total Sales ($)")

# CHART 2: Styled Horizontal Stacked Bar Chart
df_cat = (
    df.groupby(["Category", "Segment"], observed=False)["Sales_USD"]
    .sum()
    .reset_index()
)
sns.barplot(
    ax=axes[0, 1],
    data=df_cat,
    x="Sales_USD",
    y="Category",
    hue="Segment",
    palette="Purples_r",
)
axes[0, 1].set_title(
    "📊 Market Segment Distribution by Category",
    fontsize=14,
    fontweight="bold",
    color="#f8f9fa",
)
axes[0, 1].set_xlabel("Sales Volume ($)")

# CHART 3: Premium Scatter Plot with Trend Lines
sns.scatterplot(
    ax=axes[1, 0],
    data=df,
    x="Sales_USD",
    y="Profit_USD",
    hue="Category",
    palette="cool",
    s=120,
    alpha=0.8,
    edgecolor="w",
)
axes[1, 0].axhline(0, color="#ff2e63", linestyle="--", linewidth=1.5)
axes[1, 0].set_title(
    "📈 Net Profit Efficiency Margin Matrix", fontsize=14, fontweight="bold", color="#f8f9fa"
)
axes[1, 0].set_xlabel("Sales Pitch Value ($)")
axes[1, 0].set_ylabel("Net Profit ($)")

# CHART 4: Box Plot for Distribution Outliers Analysis
sns.boxplot(
    ax=axes[1, 1],
    data=df,
    x="Category",
    y="Profit_USD",
    hue="Category",
    palette="pastel",
    legend=False,
)
axes[1, 1].set_title(
    "🎯 Profit Margin Dispersion & Outliers", fontsize=14, fontweight="bold", color="#f8f9fa"
)
axes[1, 1].set_ylabel("Profit Range ($)")

# Tight Clean Optimization
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Auto-save image directly into environment workspace
output_name = "CodeAlpha_Task3_Dashboard.png"
plt.savefig(output_name, dpi=300, facecolor=fig.get_facecolor(), edgecolor="none")
plt.show()

print(f"\n🎉 Success! Super premium dashboard saved as '{output_name}'")
```
